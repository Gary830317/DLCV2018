```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap
import matplotlib.colors as colors
from sklearn.cluster import KMeans
```

P2(1)

```python
def image2feature(image):
    shape=image.shape
    image_ = image.reshape((shape[0]*shape[1],shape[2]))
    print(image_.shape)
    image_feature=[]
    for j in range(image_.shape[0]):
        a=[]
        for i in range(image_.shape[1]):    
            a.append(image_[j][i])
        image_feature.append(a)
    print(np.array(image_feature).shape)
    return image_feature
```

```python
images_names = ['mountain', 'zebra']
x1=[]
for i in range(2):
    image = np.array(cv2.imread('hw2_data/Problem2/'+images_names[i]+'.jpg', True))/255
    #cv2.namedWindow("Image")  
    #cv2.imshow('Image',image)
    #cv2.waitKey (0) 
    #cv2.destroyAllWindows()  
    shape = np.array(image[:,:,0]).shape
    x = image2feature(image)
    x1.append(x)
    kmeans = KMeans(n_clusters=10, random_state=0, max_iter=1000).fit(x)
    reconstruct_image =  np.array(kmeans.labels_).reshape(shape)
    plt.imshow(reconstruct_image, cmap="nipy_spectral")
    plt.savefig('{}_rbg.jpg'.format(images_names[i]), bbox_inches='tight', dpi=120)
```

```python
for i in range(2):
    image = cv2.imread('hw2_data/Problem2/'+images_names[i]+'.jpg', True)
    lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
    shape = np.array(lab_image[:,:,0]).shape
    x = image2feature(lab_image)
    kmeans = KMeans(n_clusters=10, random_state=0, max_iter=1000).fit(x)
    reconstruct_image =  np.array(kmeans.labels_).reshape(shape)
    plt.imshow(reconstruct_image, cmap="nipy_spectral")
    plt.savefig('{}_lab.jpg'.format(images_names[i]), bbox_inches='tight', dpi=120)
```

```python
import scipy.io as sio 
```

```python
data=sio.loadmat('hw2_data/Problem2/filterBank.mat')
print(np.array(data['F']).shape)
filter_bank = np.array(data['F'])
```

```python
x2=[]
for i in range(2):
    image = cv2.imread('hw2_data/Problem2/'+images_names[i]+'.jpg', True)
    gray_image = np.array(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY))/255
    shape = gray_image.shape
    img_feature_=[]
    for j in range(38):
        #print(filter_bank[:,:,j].shape)
        feature_map = cv2.filter2D(gray_image,-1, filter_bank[:,:,j].reshape((49,49)))
        #print(feature_map.shape)
        img_feature_.append(feature_map)  
    img_feature = np.rollaxis(np.array(img_feature_), 0, 3)
    #print(img_feature.shape)
    #plt.subplot(1,2,1),plt.imshow(gray_image, 'gray')
    #plt.subplot(1,2,2),plt.imshow(img_feature[10], 'gray')
    x = image2feature(np.array(img_feature))
    x2.append(x)
    kmeans = KMeans(n_clusters=6, random_state=0, max_iter=1000).fit(x)
    reconstruct_image =  np.array(kmeans.labels_).reshape(shape)
    plt.imshow(reconstruct_image, cmap="jet")
    plt.savefig('{}_texture.jpg'.format(images_names[i]), bbox_inches='tight', dpi=120)    
```

```python
from sklearn.preprocessing import scale

for t in range(2):
    xxx=[]
    for s in range(len(x1[t])):
        xx=[]
        xx.extend(x1[t][s])
        xx.extend(x2[t][s])
        xxx.append(xx)
    xxx = scale(xxx, axis=0)
    kmeans = KMeans(n_clusters=6, random_state=0, max_iter=1000).fit(xxx)
    reconstruct_image =  np.array(kmeans.labels_).reshape(-1, 640)
    plt.imshow(reconstruct_image, cmap="jet")
    plt.savefig('{}_all.jpg'.format(images_names[t]), bbox_inches='tight', dpi=120)   
```
